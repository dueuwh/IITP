```python
# /content/drive/MyDrive/testCode/EMMA/Experiment/IITP_TEST/Senior_Data_Excel/231221/CTB_0922/arima_rppg_databpm.csv

import csv

filename = '/content/drive/MyDrive/testCode/EMMA/Experiment/IITP_TEST/Senior_Data_Excel/231221/HJR_0301/arima_rppg_databpm.csv'  # CSV 파일의 경로
data = []

with open(filename, 'r', encoding='utf-8') as file:
    csv_reader = csv.reader(file)

    # 각 행(row)에 대하여 반복
    for row in csv_reader:
      data.append(row)
      #print(row)  # 현재 행 출력
```

```python
# /content/drive/MyDrive/testCode/EMMA/Experiment/IITP_TEST/CTB_0922.txt
filename = '/content/drive/MyDrive/testCode/EMMA/Experiment/IITP_TEST/HJR_0301.txt'  # 텍스트 파일의 경로
label_data = []

with open(filename, 'r', encoding='utf-8') as file:
    for line in file:
        #print(line.strip())  # strip() 함수로 줄 바꿈 문자 제거 후 출력
        label_data.append(line.strip().split('\t')[0])
```

```python
label_data_float = [float(i) for i in label_data]
len(label_data)
```

```python
import plotly.graph_objects as go
import numpy as np

fig = go.Figure()

fig.add_trace(go.Scatter(y=label_data_float, mode='lines', name='rppg'))

# 그래프 제목과 축 이름 설정
fig.update_layout(title='2000 Data Points: Data 1 and Data 2 Overlaid',
                  xaxis_title='Data Point Index',
                  yaxis_title='Value')

# 그래프 보기
fig.show()
```

```python
import numpy as np
import scipy.signal


def estimate_hr_from_short_ppg(ppg_samples):

    peaks, _ = scipy.signal.find_peaks(ppg_samples, distance=10)

    peaks_per_second = len(peaks)

    bpm = peaks_per_second * 60

    return bpm

estimated_heart_rates = []

for i in range(0, len(label_data), 30):
  ppg_samples = label_data[i-30:i]

  # 심박수 추정
  estimated_bpm = estimate_hr_from_short_ppg(ppg_samples)
  #print(f"Estimated BPM: {estimated_bpm}")
  estimated_heart_rates.append(estimated_bpm)
```

```python
len(estimated_heart_rates)
```

```python
data_for_plot = data_float[:len(estimated_heart_rates)]
```

```python
import plotly.graph_objects as go
import numpy as np

data1 = estimated_heart_rates
data2 = data_for_plot

# Plotly를 사용하여 그래프 그리기
fig = go.Figure()

# 데이터 1 추가
fig.add_trace(go.Scatter(y=data1, mode='lines', name='ppg2hr'))

# 데이터 2 추가
fig.add_trace(go.Scatter(y=data2, mode='lines', name='rppg2hr'))

# 그래프 제목과 축 이름 설정
fig.update_layout(title='2000 Data Points: Data 1 and Data 2 Overlaid',
                  xaxis_title='Data Point Index',
                  yaxis_title='Value')

# 그래프 보기
fig.show()
```
