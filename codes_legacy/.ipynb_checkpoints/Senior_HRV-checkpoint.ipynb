```python
import numpy as np
from scipy.signal import butter, lfilter, filtfilt
from scipy.fft import fft
from scipy.fft import fftfreq

def butter_bandpass(lowcut, highcut, fs, order=5):
    nyq = 0.5 * fs
    low = lowcut / nyq
    high = highcut / nyq
    b, a = butter(order, [low, high], btype='band')
    return b, a

def butter_bandpass_filter(data, lowcut, highcut, fs, order=5):
    b, a = butter_bandpass(lowcut, highcut, fs, order=order)
    y = filtfilt(b, a, data)
    return y

signal = tmp_list
fs = 30.0       # 샘플링 레이트 (Hz)
lowcut = 0.75   # 심박수 신호의 최소 주파수 (Hz)
highcut = 2.5   # 심박수 신호의 최대 주파수 (Hz)

# 신호 필터링
filtered_signal = butter_bandpass_filter(signal, lowcut, highcut, fs, order=6)

# FFT 적용
fft_result = fft(filtered_signal)
fft_magnitude = np.abs(fft_result)
freqs = fftfreq(len(fft_result), 1/fs)
```

```python
import matplotlib.pyplot as plt

# 그래프로 결과 확인
plt.figure(figsize=(12, 6))

# 원본 신호 그래프
plt.subplot(2, 1, 1)
plt.plot(signal, label='Original Signal')
plt.title('Original Signal')
plt.legend()

# 필터링된 신호 그래프
plt.subplot(2, 1, 2)
plt.plot(filtered_signal, label='Filtered Signal')
plt.title('Filtered Signal')
plt.legend()

plt.tight_layout()
plt.show()

# FFT 결과 그래프
plt.figure(figsize=(12, 6))
plt.stem(freqs[:len(freqs)//2], fft_magnitude[:len(fft_magnitude)//2], '-.', basefmt=" ", use_line_collection=True)
plt.title('FFT of Filtered Signal')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude')
plt.xlim(0, 10)  # 심박수 관련 주파수 범위 내에 초점을 맞춤
plt.grid()
plt.show()
```

```python
def calculate_rmse(actual, predicted):
    # 실제 값과 예측 값 사이의 차이의 제곱을 계산
    differences_squared = (actual - predicted) ** 2
    # 차이의 제곱의 평균을 계산
    mean_of_differences_squared = differences_squared.mean()
    # 계산된 평균의 제곱근을 반환
    rmse_val = np.sqrt(mean_of_differences_squared)
    return rmse_val

def calculate_mae(actual, predicted):
    # 실제 값과 예측 값 사이의 차이의 절대값을 계산
    differences = np.abs(actual - predicted)
    # 차이의 절대값의 평균을 계산
    mae_val = differences.mean()
    return mae_val

actual_values = np.array(data1)
predicted_values = np.array(data2)

# RMSE 계산
rmse = calculate_rmse(actual_values, predicted_values)
print(f"RMSE: {rmse}")

# MAE 계산
mae = calculate_mae(actual_values, predicted_values)
print(f"MAE: {mae}")
```

```python
def fill_missing_with_previous(list_, missing_value=float('nan')):
    """
    리스트에서 누락된 값을 이전 값으로 채웁니다.
    첫 번째 요소가 누락된 값일 경우, 변경하지 않습니다.

    매개변수:
    list_ (list): 처리할 리스트.
    missing_value (any, optional): 누락된 값으로 간주할 값. 기본값은 float('nan').

    반환:
    list: 누락된 값이 이전 값으로 채워진 새 리스트.
    """
    if not list_:  # 리스트가 비어있으면 그대로 반환
        return list_

    new_list = list_.copy()  # 원본 리스트를 변경하지 않도록 복사본을 생성
    for i in range(1, len(new_list)):
        # math.isnan() 함수를 사용하여 누락값이 nan인지 확인
        if math.isnan(new_list[i]):
            new_list[i] = new_list[i-1]  # 이전 값으로 누락값을 대체

    return new_list
```

```python
def calculate_rmse(actual, predicted):
    # 실제 값과 예측 값 사이의 차이의 제곱을 계산
    differences_squared = (actual - predicted) ** 2
    # 차이의 제곱의 평균을 계산
    mean_of_differences_squared = differences_squared.mean()
    # 계산된 평균의 제곱근을 반환
    rmse_val = np.sqrt(mean_of_differences_squared)
    return rmse_val

def calculate_mae(actual, predicted):
    # 실제 값과 예측 값 사이의 차이의 절대값을 계산
    differences = np.abs(actual - predicted)
    # 차이의 절대값의 평균을 계산
    mae_val = differences.mean()
    return mae_val

actual_values = np.array(data1_filtered_signal)
predicted_values = np.array(data2_filtered_signal)

# RMSE 계산
rmse = calculate_rmse(actual_values, predicted_values)
print(f"RMSE: {rmse}")

# MAE 계산
mae = calculate_mae(actual_values, predicted_values)
print(f"MAE: {mae}")
```

```python
TrainX = []
for i,j in zip(rppg_bpm_list, data2_filtered_signal):
  #print(i,j)
  TrainX.append([i,j])
```

```python
import numpy as np

# (1596, 2) 형태의 배열 생성 (예시 데이터)
data = np.array(TrainX)  # 실제 데이터로 대체해야 함

# 1596을 560으로 나누어 각 그룹의 크기 계산
group_size = len(data) // 560

# (560, 2) 형태로 데이터 재구성
# 각 그룹의 평균을 계산하여 새로운 배열 생성
averaged_data = np.array([data[i:i+group_size].mean(axis=0) for i in range(0, len(data), group_size)])

# 결과 배열 크기 확인
print(averaged_data.shape)  # (560, 2)가 출력되어야 함

# 결과 확인
#print(averaged_data)
```

```python
TrainX = averaged_data[:560]
TrainY = [[1]*80 + [0]*80 + [1]*80 + [0]*80 + [2]*80 + [2]*80 + [2]*80]
TrainY = np.array(TrainY).reshape(560,1)

np.shape(TrainY)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

X = np.array(TrainX)
y = np.array(TrainY)

# 데이터 분할
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.7, random_state=42)

# 분류기 학습
clf = RandomForestClassifier(n_estimators=100)
clf.fit(X_train, y_train)

# 예측 및 평가
y_pred = clf.predict(X_test)
print("Accuracy:", accuracy_score(y_test, y_pred))
```

```python
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

# 예제 실제 클래스 레이블과 예측된 클래스 레이블
y_true = y_test
y_pred = y_pred

# 혼동 행렬 계산
cm = confusion_matrix(y_true, y_pred)

# 혼동 행렬 시각화
plt.figure(figsize=(6,6))
sns.heatmap(cm, annot=True, fmt="d", linewidths=0.5, square=True, cmap='Blues')
plt.ylabel('Real')
plt.xlabel('Pred')
title = 'Accuracy Score: {0}'.format(np.around(accuracy_score(y_true, y_pred), decimals=2))
plt.title(title, size = 15)
plt.show()
```
